```python
# Step 1: Install and Import Libraries
!pip install pyspark seaborn pandas

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg, max as spark_max
import seaborn as sns
import pandas as pd

# Step 2: Start Spark Session
spark = SparkSession.builder.appName("LibraryDataAnalysis").getOrCreate()

# Step 3: Load Public Dataset (e.g., Titanic) via Seaborn
titanic_df = sns.load_dataset('titanic')  # Already clean and ready

# Step 4: Convert Pandas DataFrame to Spark DataFrame
spark_df = spark.createDataFrame(titanic_df)

# Step 5: Preview Data
spark_df.printSchema()
spark_df.show(5)

# Step 6: Data Cleaning (Drop Nulls)
clean_df = spark_df.dropna()

# Step 7: Perform Analysis (Example: Average Fare by Class)
clean_df.groupBy('class').agg(
    count('*').alias('passenger_count'),
    avg('fare').alias('average_fare'),
    spark_max('fare').alias('max_fare')
).show()

# Step 8: Stop Spark
spark.stop()
```
